```python
import requests
from bs4 import BeautifulSoup

urls = ['http://www.cnscg.com/forum-650-{}.html'.format(str(i)) for i in range(1,11)]
#print(urls)
def get_movies_info(urls):
    res = requests.get(urls)
    soup = BeautifulSoup(res.text,"html.parser")
    for item in soup.select('.new'):
        if len(item.select('a'))>0:
            category = item.select('em')[0].text
            title = item.select('a')[2].text
            link = 'http://www.cnscg.com/'+ item.select('a')[2]['href']
            print(category,title,link)
for movie in urls:
    get_movies_info(movie)
```
